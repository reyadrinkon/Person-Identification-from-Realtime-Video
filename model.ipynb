```python
import numpy as np
import pandas as pd
from keras.preprocessing.image import ImageDataGenerator
import os
import random 
import cv2
import imutils
import random
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelBinarizer
from keras.utils import np_utils
from keras.models import Sequential
from keras import optimizers
from sklearn.preprocessing import LabelBinarizer
from keras import backend as K
from keras.layers import Dense, Activation, Flatten, Dense,MaxPooling2D, Dropout
from keras.layers import Conv2D, MaxPooling2D, BatchNormalization
```

```python
dir = "D:\ML\ML_project_final\images\interview"
train_data = []
val_data = []
test_Data=[]
img_size = 28
for i in os.listdir(dir):
    print(i)

    count = 0
    sub_directory = os.path.join(dir,i)
    for j in os.listdir(sub_directory):
        count+=1
        if count %10==5 or count %10==7:          
            img = cv2.imread(os.path.join(sub_directory,j),0)
            img = cv2.resize(img,(img_size,img_size))
            val_data.append([img,i])
        elif count %9==0:          
            img = cv2.imread(os.path.join(sub_directory,j),0)
            img = cv2.resize(img,(img_size,img_size))
            test_Data.append([img,i])
        else:
            img = cv2.imread(os.path.join(sub_directory,j),0)
            img = cv2.resize(img,(img_size,img_size))
            train_data.append([img,i])
            

```

```python
print(len(val_data))
print(len(train_data))
print(len(test_Data))
```

```python
random.shuffle(train_data)
random.shuffle(val_data)
random.shuffle(test_Data)
```

```python
train_X = []
train_Y = []
for features,label in train_data:
    train_X.append(features)
    train_Y.append(label)

```

```python
val_X = []
val_Y = []
for features,label in val_data:
    val_X.append(features)
    val_Y.append(label)
```

```python
test_X=[]
test_Y=[]
for features,label in val_data:
    test_X.append(features)
    test_Y.append(label)
```

```python
LB = LabelBinarizer()
train_Y = LB.fit_transform(train_Y)
val_Y = LB.fit_transform(val_Y)
test_Y= LB.fit_transform(test_Y)
```

```python
train_X = np.array(train_X)/255.0
train_X = train_X.reshape(-1,28,28,1)
train_Y = np.array(train_Y)
```

```python


val_X = np.array(val_X)/255.0
val_X = val_X.reshape(-1,28,28,1)
val_Y = np.array(val_Y)
```

```python
test_X = np.array(test_X)/255.0
test_X = val_X.reshape(-1,28,28,1)
test_Y = np.array(test_Y)
```

```python


model = Sequential()

model.add(Conv2D(32, (3, 3), padding = "same", activation='relu', input_shape=(28,28,1)))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(3, activation='softmax'))

```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer="adam",metrics=['accuracy'])
```

```python
history = model.fit(train_X,train_Y, epochs=20, batch_size=32, validation_data = (val_X, val_Y),  verbose=1)
```
